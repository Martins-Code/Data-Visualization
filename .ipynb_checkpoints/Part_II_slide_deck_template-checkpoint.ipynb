## Investigation Overview

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df = pd.read_csv('clean_df')
````

> Note that the above cells have been set as "Skip"-type slides. That means
that when the notebook is rendered as http slides, those cells won't show up.



## Trip length for each user category on a weekday

> Observations on the charts below:
> - In comparison to their trips on other weekdays, both User types' weekend trips last less time.
> - The days with the highest activities seem to be Wednesday and Thursday.

````python
df_weekday = df[df.duration_min <= 60]
weekday_users = df_weekday.groupby(['user_type', 'day_of_the_week']).duration_min.mean()
weekday_users = weekday_users.reset_index()

weekday_labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
````

````python
# Convert the start_time_dayofweek to ordinal variables.
ordered_weekdays = pd.api.types.CategoricalDtype(ordered = True, categories = weekday_labels)
df['day_of_the_week'] = df['day_of_the_week'].astype(ordered_weekdays)
````

````python
# # Plot a line plot with colors based on user_type
plt.figure(figsize=[10,5])
sb.set_style('darkgrid')
sb.lineplot(data=weekday_users, x='day_of_the_week', y='duration_min', hue='user_type', markers='o')
plt.xticks(np.arange(0,7), weekday_labels);
plt.ylabel('Average Duration (min)');
````

## customers vs subscribers

> - The violinplots' widths represent the number of rides; the broader they are, the more rides they are related with. The violinplots' y-axis moves the durations up by minutes.

> - Overall, subscribers took fewer rides than customers, but they did so more frequently.

````python

# DF for user_type and duration = user_type_duration_df
user_type_duration_df = df.loc[:,['user_type', 'duration_min']]

# plot -
user_type_duration_df_60 = user_type_duration_df[user_type_duration_df['duration_min'] <= 60]
sb.violinplot(data=user_type_duration_df_60, x='user_type', y='duration_min');

# Add title and format it
plt.title('Distribution of Trip Durations by User Type'.title(),
               fontsize = 14)
# Add x label and format it
plt.xlabel('User Types'.title(),
               fontsize = 12)
# Add y label and format it
plt.ylabel('Duration in Minutes'.title(),
               fontsize = 12);
````

## Trips Duration by Age Groups and User Types

> - According to an analysis of user type, gender, and journey duration, consumers and subscribers of other genders tend to have longer trips.

````python

plt.figure(figsize = [8, 8])

ax = sb.barplot(x='user_type', y = 'duration_min', hue='member_gender', data=df.sort_values(by= 'member_gender'))
plt.title('Trips duration across user types and age groups')
plt.xlabel('User Type')
plt.ylabel('Duration (min)')
leg = ax.legend()
````
